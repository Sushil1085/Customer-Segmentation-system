```python
x = df.iloc[:,[3,4]].values
```

```python
x
```

```python
from sklearn.cluster import KMeans
WCSS =[]
```

```python
for i in range(1,11):
    kmeans = KMeans(n_clusters= i,init='k-means++',random_state=0)
    kmeans.fit(x)
    WCSS.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), WCSS)
plt.title('The elbow Method')
plt.xlabel('No of clusters')
plt.ylabel('Wcss values')
plt.show()
```

```python
k = KMeans(n_clusters = 5, init= 'k-means++',random_state=0)
```

```python
yk = k.fit_predict(x)
```

```python
plt.scatter(x[yk ==0,0],x[yk ==0,1], s =80,c = 'red',label='Customer 1')
plt.scatter(x[yk ==1,0],x[yk ==1,1], s =80,c = 'blue',label='Customer 2')
plt.scatter(x[yk ==2,0],x[yk ==2,1], s =80,c = 'yellow',label='Customer 3')
plt.scatter(x[yk ==3,0],x[yk ==3,1], s =80,c = 'cyan',label='Customer 4')
plt.scatter(x[yk ==4,0],x[yk ==4,1], s =80,c = 'black',label='Customer 5')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s =100,c='magenta',label ='centroids')
plt.title('Cluster of customer')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending score (1-100)')
plt.legend()
plt.show()
 
```
